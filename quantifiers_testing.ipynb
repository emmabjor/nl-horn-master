```python
def replace_sub_quantifiers(q_dict):
    new_dict = q_dict.copy()
    for key, value in new_dict.items():
        for sub_key, sub_value in new_dict.items():
            parenthesis_val = value[:2]+"("+value[2:]
            if sub_value == value or parenthesis_val == value:
                continue
            elif value in sub_value:
                new = sub_value.replace(value, ("*"+str(key)+"*"))
                new_dict[sub_key] = new
            elif parenthesis_val in sub_value:
                new = sub_value.replace(parenthesis_val, ("*"+str(key)+"*"))
                new_dict[sub_key] = new
    return new_dict

# replaced_dict = replace_sub_quantifiers(sorted_quantifiers)
# replaced_dict
```
